```python
from lightgbm import LGBMClassifier
%load_ext autoreload
%autoreload 2
import pandas as pd
from sklearn.externals import joblib
from sklearn.model_selection import StratifiedKFold
import xgboost as xgb
from datetime import datetime

import sys
sys.path.insert(0, 'G:/work/GitHub/ml_baseline_things/functions/')
import GridSearchCVcustom as GSC
```

```python
#load data
dataTrain, dataTarget, _ , dataTrainScaled, _ = joblib.load('data_modified/dataModified.pkl')

fileName = 'selectedFeaturesXGB-6features--2018-11-23--20-27.pkl'
selectedFeatures = joblib.load('selectedFeatures/'+ fileName)

dataTrain = dataTrain[selectedFeatures]
dataTrainScaled = dataTrainScaled[selectedFeatures]
```

```python
#check available params
pipeline = GSC.get_pipeline_1step(xgb.XGBClassifier(n_jobs=-1))
pipeline.named_steps 
```

```python
RS = 100
n_splits = 5
nSamples = len(dataTrain)
cvShuffle = True
paramGrid = {'clf': [xgb.XGBClassifier(n_jobs=-1)],
              'clf__random_state': [RS],
              'clf__class_weight': ['balanced'],
              'clf__n_estimators': [10, 100, 250, 500, 750, 1000],
              'clf__max_depth': [1, 2, 4],
              'clf__learning_rate': [0.01, 0.02, 0.05, 0.1, 0.5]
             }
```

# non-Scaled + StratifiedKFold

```python
cv = StratifiedKFold(n_splits=n_splits, random_state=RS, shuffle=cvShuffle)

bestModelScore1, bestModelParams1, bestModelIter1 = GSC.GridSearchCVcustom(pipeline, paramGrid, 
                                                              dataTrain[:nSamples], dataTarget[:nSamples],
                                                              cv=cv, n_jobs = -1, pre_dispatch=2,
                                                              showSteps=True, itersToPerform ='All')
print ('bestScore = {}\nBestScoreParams = {}'.format(bestModelScore1, bestModelParams1))
```

# Scaled + StratifiedKFold

```python
bestModelScore2, bestModelParams2, bestModelIter2 = GSC.GridSearchCVcustom(pipeline, paramGrid, 
                                                              dataTrainScaled[:nSamples], dataTarget[:nSamples],
                                                              cv=cv, n_jobs = -1, pre_dispatch=2,
                                                              showSteps=True, itersToPerform ='All')
print ('bestScore = {}\nBestScoreParams = {}'.format(bestModelScore2, bestModelParams2))
```
